```python
import numpy as np

l = np.loadtxt('./data/diabetes.csv', delimiter=',', skiprows=1)

x = l[:, 0:-1]
y = l[:, [-1]]

print(l.shape)
print(x.shape)
print(y.shape)
print(x[0])
print(y)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Dense(12, input_shape=(x.shape[1],), activation='relu'))
model.add(Dense(x.shape[1], activation='relu'))
model.add(Dense(y.shape[1], activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
training = model.fit(x, y, epochs=150, batch_size=10, verbose=0)
```

```python
model.evaluate(x, y)
```

```python
import matplotlib.pyplot as plt

plt.xlabel('epochs')
plt.ylabel('loss')
plt.xlim(0,150)
plt.grid()

plt.plot(training.history['loss'], label='loss')
plt.legend(loc='best')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.grid()

plt.plot(training.history['accuracy'], label='accuracy')
plt.legend(loc='best')

plt.show()
```
